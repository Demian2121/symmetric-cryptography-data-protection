Proyecto de Seguridad Informática – Implementación de cifrado y descifrado de información utilizando Python

```python
# cifrar
from cryptography.hazmat.primitives.asymmetric import rsa, padding
from cryptography.hazmat.primitives import hashes

private_key_rsa = rsa.generate_private_key(public_exponent=65537, key_size=2048)
public_key_rsa = private_key_rsa.public_key()

mensaje = b'mensaje cifrado'

cifrado = public_key_rsa.encrypt(
    mensaje,
    padding.OAEP(
        mgf=padding.MGF1(algorithm=hashes.SHA256()),
        algorithm=hashes.SHA256(),
        label=None
    )
)

print("texto cifrado",cifrado.hex())
```

```python
# decifrar
decifrado = private_key_rsa.decrypt(
    cifrado,
    padding.OAEP(
        mgf=padding.MGF1(algorithm=hashes.SHA256()),
        algorithm=hashes.SHA256(),
        label=None
    )
)
print("texto decifrado",decifrado.decode("utf-8"))



```

```python
# import DSA para firma digital
from cryptography.hazmat.primitives.asymmetric import dsa
```

```python
# generar la clave
private_key_dsa = dsa.generate_private_key(key_size=2048)
public_key_dsa = private_key_dsa.public_key()
```

```python
# firmar el mensaje
firma = private_key_dsa.sign(
    mensaje,
    hashes.SHA256()
)
```

```python
# verificar firma
try:
    public_key_dsa.verify(
        firma,
        mensaje,
        hashes.SHA256()
    )
    print("firma valida")
except:
    print("firma invalida")
```
